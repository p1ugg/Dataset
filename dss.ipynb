````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
BOOKS = pd.read_excel('../data.xlsx')
````

### Графический отчет «кластеризованная столбчатая диаграмма» для пары «качественный атрибут —качественный атрибут»

````python
data = {
    'Country': BOOKS['Country'].tolist(),
    'Genre': BOOKS['Genre'].tolist()
}

df = pd.DataFrame(data)

count_data = df.groupby(['Country', 'Genre']).size().unstack()

# Создание кластеризованной столбчатой диаграммы
count_data.plot(kind='bar', stacked=False)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Clustered Bar Chart for Country and Genre')
plt.legend(title='Genre')
plt.show()
````

````python
data = {
    'Genre': BOOKS['Genre'].tolist(),
    'Country': BOOKS['Country'].tolist()
}

df = pd.DataFrame(data)

count_data = df.groupby(['Genre', 'Country']).size().unstack()

# Создание кластеризованной столбчатой диаграммы
count_data.plot(kind='bar', stacked=False)
plt.xlabel('Genre', fontsize=2)
plt.ylabel('Country', fontsize=2)
plt.title('Clustered Bar Chart for Genre and Country')
plt.legend(title='Country')
plt.legend(fontsize=7)
plt.show()
````

###Графический отчет «категоризированная гистограмма» для пары «количественный атрибут—качественный атрибут»

````python
data = {
    'User Rating': BOOKS['User Rating'].tolist(),
    'Genre': BOOKS['Genre'].tolist()
}

df = pd.DataFrame(data)
# Группировка данных по категориальному атрибуту и построение гистограммы
plt.figure(figsize=(8, 6))
for category in df['Genre'].unique():
    subset = df[df['Genre'] == category]
    plt.hist(subset['User Rating'], alpha=0.5, label=category)

plt.xlabel('User Rating')
plt.ylabel('Frequency')
plt.title('Categorized Histogram for User Rating by Genre')
plt.legend(title='Genre')
plt.show()
````
